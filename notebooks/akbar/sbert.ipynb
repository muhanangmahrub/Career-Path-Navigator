ðŸŸ¦ Cell 8 â€” Ranking Job

```python
ranking = sorted(
    zip(job_names, scores),
    key=lambda x: x[1],
    reverse=True
)

ranking
```

```python
print("=== REKOMENDASI JOB ===")
for job, score in ranking:
    print(f"{job:30s} -> {score:.2f}")

top_job = ranking[0][0]
print(f"\nðŸŽ¯ Job paling cocok: {top_job}")
```
