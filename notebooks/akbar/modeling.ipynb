```python
import re
import pandas as pd

edu_map = {
    r'SMA|SMU|SMK|STM': ['sma', 'smk'],
    # r'Sertifikat Professional': ['sertifikat'],
    r'D3 \(Diploma\)': ['d3'],
    r'D4 \(Diploma\)': ['d4'],
    r'Sarjana \(S1\)': ['s1'],
    # r'Diploma Pascasarjana': ['pascasarjana'],
    # r'Gelar Professional': ['professional'],
    r'Magister \(S2\)': ['s2'],
    r'Doktor \(S3\)': ['s3']
}

def extract_education_levels(text):
    if pd.isna(text) or text == 'Tidak terspesifikasi':
        return []

    result = []

    for pattern, labels in edu_map.items():
        if re.search(pattern, text):
            result.extend(labels)

    return sorted(set(result))
```

```python
df['education_norm'] = df['education_level'].apply(extract_education_levels)

print("Extract success")
```

```python
# prioritas kategori (urutannya penting)
priority = list(job_label.keys())

def assign_label(raw_title):
    t = clean_text(raw_title)
    t_pad = " " + t + " "  # biar cari kata utuh lebih gampang
    for cat in priority:
        for kw in job_label[cat]:
            if kw in t_pad:
                return cat
    return "Other"

labels = [assign_label(t) for t in df['job_title']]
```

```python
LABEL_KEYWORDS = {
    "Data": [
        "data", "analyst", "analytics", "scientist", "machine learning", "ml", "ai"
    ],
    "Tech": [
        "developer", "programmer", "software", "engineer", "backend",
        "frontend", "fullstack", "devops", "cloud", "it ", "it-"
    ],
    "Marketing": [
        "marketing", "digital marketing", "seo", "content",
        "brand", "promotion", "growth"
    ],
    "Sales": [
        "sales", "account executive", "business development",
        "relationship", "key account", "bd"
    ],
    "Finance": [
        "finance", "accounting", "accountant", "auditor",
        "tax", "treasury", "financial"
    ],
    "HR": [
        "hr", "human resource", "recruiter",
        "talent", "people development"
    ],
    "Operations": [
        "operation", "operations", "logistic", "warehouse",
        "supply chain", "procurement", "admin", "staff"
    ],
    "Design": [
        "designer", "ui", "ux", "graphic", "visual"
    ],
    "Education": [
        "teacher", "lecturer", "instructor", "tutor"
    ],
    "Healthcare": [
        "doctor", "medical", "nurse", "clinic", "hospital"
    ]
}
```

```python
def classify_job(title):
    if not isinstance(title, str):
        return "Other"

    title = title.lower()

    for label, keywords in LABEL_KEYWORDS.items():
        for kw in keywords:
            if kw in title:
                return label

    return "Other"

df['label2'] = df['job_title'].apply(classify_job)
```
